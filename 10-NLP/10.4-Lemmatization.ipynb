- Lemmatization

Lemmatization is an advanced text normalization technique that reduces a word to its true dictionary base form, which is called a lemma.

Unlike stemming - which blindly chops off prefixes or suffixes using structural rules - lemmatization uses a dictionary look-up and analyzes the vocabulary and grammar rules of a language to return a valid word.

NLTK provides the WordNetLemmatizer class. 
To use it, I first need to download the WordNet database asset:

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download the required dictionary database
# nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("histories"))  
print(lemmatizer.lemmatize("history"))    
print(lemmatizer.lemmatize("programming")) # Output: programming (Needs Part of speech (POS) tag!)

print(lemmatizer.lemmatize("programming", pos = 'v')) #verb
print(lemmatizer.lemmatize("programming", pos = 'a')) # adjective
print(lemmatizer.lemmatize("programming", pos = 'r')) # adverb
```

Can be use in chatbot, QA, text summarization
