### Named Entity Recognition (NER)

Named Entity Recognition (NER) is an advanced NLP technique that scans unstructured text to automatically find and classify specific real-world proper nouns into predefined categories like People, Companies, Locations, Dates, or Monetary Values.

If Part-of-Speech (POS) tagging identifies the grammar skeleton (like "this is a noun"), NER goes a step further and identifies the identity of that noun (like "this noun is Elon Musk").


#### eg - 

Raw Input: "Sundar Pichai visited the Google headquarters in California in 2024.

#### NER Output:

- Sundar Pichai -> PERSON
- Google -> ORGANIZATION (or ORG)
- California -> GPE (Geopolitical Entity / Location)
- 2024 -> DATE

```python
import nltk
from nltk.tokenize import word_tokenize

text = "Elon Musk discussed SpaceX launching its new rocket from Texas to explore the Mars equator."

# Download once
# nltk.download('maxent_ne_chunker_tab')
# nltk.download('words')

# Tokenize first
tokens = word_tokenize(text)

# pos tag (NER relies heavily on grammar tags!)
pos_tags = nltk.pos_tag(tokens)

# perform named entity recognition
# ner_tree = nltk.ne_chunk(pos_tags, binary = True)
ner_tree = nltk.ne_chunk(pos_tags).draw()
ner_tree = nltk.ne_chunk(pos_tags)

print(ner_tree)
```

Pos tag and NER are doing the exact same thing because both processes assign labels to words, and both labels look similar (like NNP).

However, they are fundamentally different tasks. POS Tagging tells you the word's grammatical type, while NER tells you the word's real-world identity.
