- Word2Vec

Word2Vec is a technique that converts words into dense vectors (embeddings) such that words with similar meanings have similar vector representations.



```python
import gensim
from gensim.models import word2vec, KeyedVectors

import numpy as np
import pandas as pd

from nltk import sent_tokenize
from gensim.utils import simple_preprocess

sentences = [
    ["king", "is", "a", "man"],
    ["queen", "is", "a", "woman"],
    ["prince", "is", "a", "young", "king"],
    ["princess", "is", "a", "young", "queen"],
    ["man", "and", "woman"],
    ["king", "and", "queen"]
]

# Model configuration
model = gensim.models.Word2Vec(
    sentences=sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=1
)

# Vocabulary
model.build_vocab(sentences)

# train model
model.train(sentences, total_examples=model.corpus_count, epochs=model.epochs)

print(model.wv.most_similar('king'))
print(model.wv.doesnt_match(['king', 'queen', 'man', 'prince']))
print(model.wv['king'])

print(model.wv.similarity('king', 'prince'))

y = model.wv.index_to_key
print(y)

# The reason the results look strange is because my training data is extremely small.
```
