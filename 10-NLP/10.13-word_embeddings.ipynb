- Word Embeddings 

Word embedding is a modern NLP technique that converts words into dense vectors of real numbers. It maps words into a multi-dimensional continuous vector space, capturing semantic meaning, context, and relationships between words.

Unlike older keyword-matching methods like TF-IDF or Bag of Words, word embeddings allow machines to understand that "king" and "queen" or "coffee" and "tea" are conceptually related, even if they share zero characters in common.
