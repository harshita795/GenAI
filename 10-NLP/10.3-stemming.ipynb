# Stemming

Stemming is the process of cutting off the ends of words to reduce them to their base or root form (called a stem).
It is a crude, rule-based text normalization technique used in NLP to make sure the computer treats different variations of the same word as the single identical concept.

- Why do we use it?
  - Reduces Vocabulary Size: Instead of your machine learning model learning 4 separate words (dance, dances, dancing, danced), it only has to learn 1 token (danc).

  - Improves Search Engines: If a user searches for "running shoes", a stemming-backed system will match it to documents containing "run shoes" or "ran shoes".

### Popular Stemmers in NLTK
NLTK provides two major stemming algorithms:
- **Porter Stemmer**: The oldest and most gentle stemming algorithm. It uses a standard set of structural rules.


- **Lancaster Stemmer**: A much more aggressive stemmer. It chops words down heavily, often leaving roots that are very difficult for humans to read.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

print(stemmer.stem("programming")) 
print(stemmer.stem("programs"))    
print(stemmer.stem("programmed"))   
print(stemmer.stem("fairly")) 
print(stemmer.stem("generously"))
    
print(stemmer.stem("history")) # it does not care about word's meaning, it just stem it, and for this case Lemmetization works which provide meaning.
```

- RegexpStemmer 

A RegexpStemmer (Regular Expression Stemmer) is a specialized NLTK tool that lets you build your own custom stemmer by defining exact pattern-matching rules using regular expressions.

```python
from nltk.stem import RegexpStemmer

# This custom stemmer removes 'ing', 's', 'ed', or 'able' from the end of words
# But only if the word has at least 4 characters
custom_stemmer = RegexpStemmer('ing$|s$|ed$|able$', min=4)

# The dollar sign ($) ensures it only strips letters from the end of a word.

print(custom_stemmer.stem("cooking"))     # Output: cook  (removes 'ing')
print(custom_stemmer.stem("dogs"))        # Output: dog   (removes 's')
print(custom_stemmer.stem("advisable"))   # Output: advis (removes 'able')
print(custom_stemmer.stem("beds"))        # Output: bed   (removes 's')
print(custom_stemmer.stem("was"))         # Output: was   (IGNORED because len("was") < 4)
```


- Snowball Stemmer

The Snowball Stemmer is a significantly improved, faster, and more aggressive version of the classic Porter Stemmer

- Imporvements are - 
  - Multi-Language Support:
  - Better Speed & Efficiency
  - Handles Quirky Rules Better

```python
from nltk.stem import SnowballStemmer

# Initialize the Snowball Stemmer for English
snowball = SnowballStemmer(language='english')

print(snowball.stem("programming"))  # Output: program
print(snowball.stem("fairly"))       # Output: fair
print(snowball.stem("generously"))   # Output: generous
print(snowball.stem("history"))

```
