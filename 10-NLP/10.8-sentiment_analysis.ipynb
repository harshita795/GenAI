- Sentiment Analysis

"I absolutely love this product! The battery life is amazing." -> Positive (+0.9)
- Input 2: 

"The shipping took two weeks and the screen arrived broken." -> Negative (-0.8)
- Input 3: 

"The package includes a charging cable and a user manual." -> Neutral (0.0)
