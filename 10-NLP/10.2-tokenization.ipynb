# Tokenization in NLP
- Tokenization is the process of breaking down sensitive data or raw information into smaller, distinct, and manageable units called "tokens"


- Corpus (paragraph)
- Documents (sentences)
- Vocubulary (Unique words)
- Words (specific word)

### Tokenization
eg - text -> 
("My name is Harshita and I like to study hard any concept. I am Happy.") - *this paragraph is corpus*

- From this paragraph we generate sentences, hence that sentence called token, and process called tokenization.
  - We will get 2 sentences from that paragraph.

  - document 1 -> My name is Harshita and I like to study hard any concept 
  - document 1 -> I am Happy

  
- We can also perform tokenization on these each document into words.

### Vocabulary
Example -
- I like to dance. My friend likes to play.

after performing tokenization, we get - 

- I like to dance
- My friend like to play

Total I have 9 words here.
Unique words -> 7 (like, to -> duplicated words)

These 9 words we have in vocabulary.

```python
# Tokenization Example
corpus = """Hello welcome, my name is Harshita. 
I am learning NLP here.
I am trying to understand's it deeply!
I can't waste time.
I am in U.S.A.
"""
print(corpus)
```

```python
# import nltk

# Download the required tokenization models
# nltk.download('punkt')
# nltk.download('punkt_tab')

# Need to install these data model to do tokenization in nltk
```

```python
# Convert paragraphs into  senctences 
from nltk.tokenize import sent_tokenize

print(sent_tokenize(corpus))
```

```python
documents = sent_tokenize(corpus)
type(documents)

for sentence in documents:
  print(sentence)
```
