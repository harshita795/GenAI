- TF-IDF

TF-IDF stands for Term Frequency-Inverse Document Frequency. 
TF-IDF converts text into numerical vectors. 

It is a scoring system used by computers to figure out how important a word is inside a specific document compared to a whole collection of documents.



## The 2 Core Parts (How it is Built)

### 1. Term Frequency (TF) - *The Local Counter*

- **What it is:** How many times a word appears in a single document.

- **Formula (Simple):**

$$
\text{TF} = \frac{\text{Count of the word}}{\text{Total words in that document}}
$$

- **Logic:** If a word appears a lot, it is highly relevant to *this specific text*.

---

### 2. Inverse Document Frequency (IDF) - *The Rarity Scale*

- **What it is:** How rare or common a word is across your entire library of documents.

- **Formula (Simple):**

$$
\text{IDF} = \log\left(\frac{\text{Total number of documents}}{\text{Number of documents containing the word}}\right)
$$

- **Logic:** Common words (like **"the"**, **"is"**) get heavily penalised. Rare words (like **"cybersecurity"**, **"recipe"**) get a major boost.



## The Final Math Formula

### TF-IDF Score - *The Importance Metric*

- **Formula:**

$$
\text{TF-IDF Score} = \text{TF} \times \text{IDF}
$$

- **High Score:** Word is frequent in this document, but rare overall  
  *(e.g., "Quantum" in a physics paper).*

- **Zero Score:** Word appears in every single document  
  *(e.g., "And", "Of").*

- **Logic:** TF-IDF rewards words that are highly relevant to a specific document while reducing the importance of words that are common across many documents.

---






## The Ultimate Example (The 3-Book Library)

Imagine a library with exactly **3 books**:

- **Book 1:** About Pizza 
- **Book 2:** About Pasta 
- **Book 3:** About Burgers 

Let's calculate the score for two words inside **Book 1 (Pizza)**.

---

### Word A: "The"

- **TF:** Appears 100 times in Book 1 *(High)*.
- **IDF:** Appears in all 3 books

$$
\frac{3}{3} = 1
\qquad\Rightarrow\qquad
\log(1) = 0
$$

- **TF-IDF Score:**

$$
100 \times 0 = \mathbf{0}
$$

- **Result:** The computer ignores it.

---

### Word B: "Pepperoni"

- **TF:** Appears 20 times in Book 1 *(High)*.
- **IDF:** Appears in only 1 out of 3 books *(High rarity boost)*.

- **TF-IDF Score:**

$$
20 \times \text{High Rarity}
=
\mathbf{\text{Huge Score}}
$$

- **Result:** The computer flags Book 1 as specifically about **"Pepperoni"**.

---

### Key Insight

TF-IDF is designed to answer one simple question:

> **"Which words make this document unique compared to all the others?"**

Common words get a score close to **0**, while distinctive words receive a **high score** and become powerful signals for search engines, recommendation systems, and document ranking algorithms.
