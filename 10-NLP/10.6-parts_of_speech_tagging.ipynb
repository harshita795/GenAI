- Part-of-Speech
 
Part-of-Speech (POS) tagging is the process of reading a text and automatically labeling each word with its correct grammatical category - such as Noun, Verb, Adjective, or Adverb - based on both its definition and its context within the sentence.

In NLP, a computer cannot understand a sentence just by looking at raw words. POS tagging gives the computer the grammatical context it needs to figure out what the words are actually doing.

Think of POS Tagging as a Sensor and Lemmatizer as a Tool.

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the tagger asset (only needed once)
# nltk.download('averaged_perceptron_tagger_eng')

sentence = "The quick brown fox jumps over the lazy dog."

words = word_tokenize(sentence)

tagged_words = nltk.pos_tag(words)

print(tagged_words)
```

These short codes are called POS Tags (Part-of-Speech Tags). Specifically, they come from the Penn Treebank Tagset, which is the standard system used in English Natural Language Processing to categorize grammar.


### Penn Treebank POS tags


1. Nouns (The Core Subjects)
- NN: Noun, singular or mass (e.g., computer, sand, air)
- NNS: Noun, plural (e.g., computers, tables, cities)
- NNP: Proper noun, singular (e.g., London, Google, John)
- NNPS: Proper noun, plural (e.g., Americans, Mondays, Pyramids)
