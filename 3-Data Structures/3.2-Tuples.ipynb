Explanation:

- Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(lst)
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
# list to tuple conversion

numbers = tuple([1, 2, 3, 4, 5, 6])
numbers
```

```python
# tuple to list conversion
list((1,2,3,4,5))
```

```python
mixed_tuple = (1, "hello", 4.5, True)
print(mixed_tuple)
```

```python
# Accessing Tuple elements
print(numbers)
print(numbers[2])
print(numbers[-1])
```

```python
# slicing in tuple
print(numbers[0:4])
print(numbers[::-1])
```

```python
# Tuple operations

concat_tuple = numbers + mixed_tuple
print(concat_tuple)

multiple_tuple = mixed_tuple * 3
print(multiple_tuple)
```

```python
# Immutable nature of tuples

lst = [1, 2, 3, 4, 5]
print(lst)

lst[1] = "Krish"
print(lst) # list are mutable

tple = (1, 2, 3, 4, 5)
print(tple)

tple[1] = "Krish"
print(tple) # tuples are immutable, cannot change element once assigned
```

```python
# Tuple methods
print(numbers)
print(numbers.count(2))
print(numbers.index(1))
```

```python
# Packing and Unpacking Tuples
packed_tuple = 1, "Hello", 3.24
print(packed_tuple) # It pack them in tuple dataType

a,b,c = packed_tuple # unpacking the tuple elements to the variables a,b,c respectively
print(a)
print(b)
print(c)
```

```python
# Unpacking with *
print(numbers)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
# Nested tuple
lst = [[1,2,3,4], [6,7,8], [1,2]]
print(lst)
print(lst[0][2])
print(lst[0][0:3])

tple = ((1,2), (1,5,6,7), (3,4,5,7))
print(tple)
print(tple[2][3])
print(tple[1][1:5])
```

```python
#  iterating over nested tuples
for sub_tuple in tple:
  for item in sub_tuple:
    print(item, end=" ")
```
