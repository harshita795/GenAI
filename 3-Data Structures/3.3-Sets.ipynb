# Sets

Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
# create a set
my_set = {1,2,3,4,5}
print(my_set)
empty_sets = {}
print(type(my_set))
print(empty_sets)
```

```python
lst_to_set = set([1,2,3,4,5])
print(lst_to_set)
```

```python
# Basic sets operations
print(my_set)
my_set.add(6)
print(my_set)
```

```python
# removing the element from the sets
print(my_set)
my_set.remove(2)
print(my_set)
```

```python
# The main difference is how they handle missing items: remove() throws an error if the item is not found, while discard() does nothing.
removing = my_set.remove(11)
print(removing)
```

```python
discarding = my_set.discard(11)
print(discarding)
```

```python
# pop method
print(my_set)
my_set.pop()
print(my_set)
# sets do not have indexes or positions. Because there is no "index 2" in a set, passing a number inside .pop() is illegal. It accept no arguments.
my_set.pop(2)
print(my_set)
```

```python
my_set.clear()
print(my_set)
```

```python
# set membership test
my_set = {1,2,3,4,5}
print(3 in my_set)
print(6 in my_set)
```

```python
# Mathematical operation
set1 = {1,2,3,4,5,6}
set2 = {7,8,9,4,5,6}

# Union
union_set = set1.union(set2)
print(union_set) 

# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set) # common elements

# When you run set1.intersection_update(set2), Python alters set1 so it only contains those overlapping items. "1,2,3" is thrown away.
# this does not create a new set,i it saves memory.
set1.intersection_update(set2)
print(set1)

# difference
set1 = {1,2,3,4,5,6}
set2 = {7,8,9,4,5,6}
difference_set = set1.difference(set2)
print(difference_set) #Creates a temporary new set with items unique to set1
differ = set1.difference_update(set2)
print(differ) # It returns NOTHING (None) to the print statement.
print(set1) # Modifies set1 in place
```

- If a set method name does not end in _update (like .difference(), .intersection(), .union()), it creates and returns a new set. You can print it directly.
- If a set method name ends in _update (like .difference_update(), .intersection_update()), it mutates the original set and returns None. You must print the variable on the next line to see the changes.

```python
# symmetric difference
set1 = {1,2,3,4,5,6}
set2 = {7,8,9,4,5,6}
set1.symmetric_difference(set2)
```

```python
# set_methods
set1 = {1,2,4}
set2 = {1,2,4,3,6}

# is_subset
print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
lst = [1,2,3,4,5]
set(lst)
```

```python
# Counting unique words in text
text = "Counting unique words in text"
words = text.split()
print(words)
set_words = set(words)
print(set_words)
```
