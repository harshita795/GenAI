1. Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
# Creating Dictionaries
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
student = {"name": "Harshita", "age": 24, "grade": 95}
print(student)
print(type(student))
```

```python
# Unique Key
new_student = {"name": "Harshita", "age": 24, "name": 95}
print(new_student) # key should be unique, it will use last value
```

```python
# Accessing Dictionary elements
print(student)
print(student["grade"])
print(student["age"])
print(student["name"])

# Acccessing using get() method
print(student.get("grade"))
print(student.get("name"))
print(student.get("age"))
print(student.get("last_name")) # None
print(student.get("last_name", "Not Available")) # default value provided
print(hash("age"))
```

```python
# Modifying Dictionary elements
# Dictionary are mutable, so you can add, update, delete elements

# updating key value
student["age"] = 25
print(student)

# added new key value
student["address"] = "India"
print(student)

# delete the key
del student["name"]
print(student)
```

```python
# Dictionary methods

keys = student.keys()
print(keys)
values = student.values()
print(values)
items = student.items()
print(items)
```

```python
# Shallow copy
student_copy = student
print(student)
print(student_copy)

student["address"] = "United States"
print(student)
print(student_copy)

student_copy1 = student.copy()

student["address"] = "United Kingdom"
print(student_copy1)
print(student)
```

```python
# Iterating over dictionaries
# Use loops to iterate over dictionaires, keys, values, items

# Iterate over keys
print(student)
for key in student.keys():
  print(key)

# iterate over values
for value in student.values():
  print(value)

# iterate over key-value
for key, value in student.items():
  print(f"{key} : {value}") 
```

```python
# Nested Dictionaries
students = {
  "student1": {"name": "Harshita", "age": 24},
  "student2": {"name": "Cutie", "age": 26}
}

# Access nested dictionaries
print(students["student2"]["name"])
print(students["student2"]["age"])

# Iterating over nested dictionaries
for student_id, student_info in students.items():
  for key, value in student_info.items():
    print(f"{key} : {value}")
```

```python
# Dictionary Comprehension
for x in range(5):
  print(f"{x}: {x**2}")
   
square = {x:x**2 for x in range(5)}
print(square) # dict comprehension

even_square = {x:x**2 for x in range(10) if x % 2 == 0}
print(even_square)

square_if_odd_cube_if_even = {x: (x**2 if x % 2 != 0 else x**3) for x in range(10) }
print(square_if_odd_cube_if_even)
```

```python
# examples
# use a dictionary to count the frequency of element in list.

numbers = [1,2,2,2,2,3,3,3,3,3]
frequency = {}
# counting_numbers = {x: numbers.count(x) for x in numbers}
# print(counting_numbers)

for number in numbers:
  if number in frequency:
    frequency[number] += 1
  else:
    frequency[number] = 1
print(frequency)

# counter tool is more efficient in python for finding frequency
```

```python
# Merge two dictionaries 
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged_dict = {**dict1, **dict2} # keyword argument
print(merged_dict)
```
