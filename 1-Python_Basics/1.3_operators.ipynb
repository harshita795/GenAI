```python
# Arithmetic operations

a = 10
b = 20
c = a + b # addition
print(c) 

c = a - b # substraction
print(c)

c = a * b # multiplication
print(c)

c = a / b # division
print(c) 

c = a % b # modulus
print(c)

c = a // b # floor
print(c)

c = a ** b # exponentiation
print(c)

```

```python
# Comparision operators
a = 20
b = 30

c = (a == b) # equal to
print(c)

c = (a != b) # not equal to
print(c)

c = (a > b) # greater than
print(c)

c = (a < b) # less than
print(c)

c = (a >= b) # greater than or equal to
print(c)

c = (c <= b) # less than or equal to
print(c)

```

```python
# Logical operators

a = True
b = False

c = a and b # AND (if both are true then will get only true)
print(c)

c = a or b # OR (first element should be true to get true)
print(c)

c = not a # NOT (it give reverse)
print(c)
```
