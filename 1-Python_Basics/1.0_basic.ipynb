```python
# Basic Syntax rules in Python

# Both these variables are different identifiers, then are not same variables, Python is case-sensitive
name = "Harshita"
Name = "Harshi"

print(name)
print(Name)
```

```python
# Indentation

# Python uses indentation to define block of code.
age = 32
if age > 30:
  print(age)
  
```

```python
# Line continuation

# use a backslash (\) to continue statement to the next line
total = 1+2+3+4+5+6+7+\
4+5+6
print(total)
```

```python
# Multiple staements on a single line

x = 4; y = 5; z = x + y
print(z)
```

```python
# Understanding Semantics in Python

# variable assignment
# kernel will understand that age is integer at runtime
age = 32 # age is integer
name = "Harshita" # name is string

# Type Inference
print(type(age))# integer
print(type(name)) # string
```
