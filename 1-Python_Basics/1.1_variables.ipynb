```python
# Declaring and assigning variables
age = 24
name = "Harshita"
is_Human = True
height = 5.4

# printing the variables
print("age :", age)
print("Height :", height)
print("is_human :", is_Human)
print("name :", name)
```

```python
# Naming Convention
# Variable names should be descriptive
# They must always start with a letter and underscore '_', and can contain letters, numbers and underscores 
# Variables names are case sensitive

# valid variable names
first_name = "Harshita"
last_name = "Yadav"
age = 24

# invalid variable names
# 12firstname = "Harshita"
# first-name = "harshita"
# @firstname = "harshita"
```

```python
# Understanding variable types
# Python is dynamically types (type of a variable is determined at runtime)

age = 25 # integer
height = 5.4 # float
name = "Harshita" # string
is_Human = True # boolean

print(type(age))
print(type(name))
print(type(is_Human))
print(type(height))
```

```python
# Type checking 
age = 24
print(type(age))
new_age = "25"

# Type conversion
age_str = str(age)
print(age_str)
print(type(age_str))
print(type(int(new_age)))
```

```python
# Dynamic typing
# Python allows the type of variable to change as the program excute

var = 10 # int
print(var, type(var))

var = "Hello" # string
print(var, type(var))

var = 3.14 # float
print(var, type(var))
```

```python
# User Input
age = int(input("What is the age?"))
print(age, type(age))
```

```python
# Calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("sum: ", sum)
print("difference: ", difference)
print("product: ", product)
print("quotient: ", quotient)
```
