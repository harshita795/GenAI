# GRU (Gated Recurrent Unit)

## Why GRU
- In LSTM there are lot of parameters  and architecture was complex, so it take more training time.
- GRU is simpler architecture than LSTM.
- LSTM has 3 gates while GRU has 2 gates, and less parameters, hence less training time.
- GRU has 1 memory while LSTM has 2 memory.
- But both work good in other applications. Both are good tools.


# Architecture

- GRU carry both long term memory and short term memory in single hidden state, it does not have cell state Ct.

- It contains only 2 gates.
  - Reset Gate
  - Update Gate





## The Vectors

- ht-1 -> Hidden state
- ht -> Current hidden state
- Xt -> Input
- rt -> Reset gate
- zt -> Update gate
- ht -> Candidate hidden state
- These all are vectors
- Other than Xt (input), all other vectors have same dimensions.
