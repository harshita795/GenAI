- File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. 

```python
# Read a whole file

with open("example.txt", "r") as file:
  content = file.read()
  print(content)
```

```python
# Read a file line by line
with open("example.txt", "r") as file:
  for line in file:
    print(line.strip()) # strip() function remove newline character
```

```python
# Writing a file(Overwriting)

with open("example.txt", "w") as file:
  file.write("Hello Guys!") # this overwrite the existing file.

with open("example.txt", "r") as file:
  print(file.read())
```

```python
# write a file without overwriting
with open("example.txt", "a") as file:
  file.write("\nAppend is taking place.")

with open("example.txt", "r") as file:
  print(file.read())
```

```python
# wrting list of lines to a file
lines = ["\nFirst line \n", "Second line \n", "Third line \n"]

with open("example.txt", "a") as file:
  file.writelines(lines)

with open("example.txt", "r") as file:
  print(file.read())
```

```python
# Binary files
data = b"\x00\x01\x02\x03\x04"
with open("example.bin", "wb") as file:
  file.write(data)

with open("example.bin", "rb") as file:
  print(file.read())
```

```python
# Read the content from a source text file and write it to a destination text file

with open("example.txt", "r") as example_file:
  content = example_file.read()

with open("destination.txt", "w") as destination_file:
  destination_file.write(content)
```

```python
# read a text file and count the number of lines , words, and characters.

with open("example.txt", "r") as file:
  lines = file.readlines()
  
  total_lines = len(lines)
  total_words = 0
  total_chars = 0
  for line in lines:
    total_chars += len(line)
    words = line.split()
    total_words += len(words)
    
print(f"total lines : {total_lines}")
print(f"total words : {total_words}")
print(f"total chars : {total_chars}")
```

```python
# writing and then reading a file
with open("example.txt", "w+") as file:
  file.write("Helo world\n")
  file.write("This is good to know\n")

  # Move the file cursor to the beginning
  file.seek(5)

  print(file.read())
```
