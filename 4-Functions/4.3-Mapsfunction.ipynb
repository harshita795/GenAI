- The map() Function in Python

The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
  return x * x

print(square(10))

# map function
numbers = [1,2,3,4,5,6]
print(list(map(square, numbers)))
```

```python
# Lambda function with map
print(list(map(lambda x: x * x, numbers)))
```

```python
# Map multiple iterables
numbers1 = [1,2,3]
numbers2 = [4,5,6]

list(map(lambda x, y: x + y, numbers1, numbers2))
```

```python
# map to convert list of string to integers
str_numbers = ["1", "2", "3", "4", "5"]
int_numbers = list(map(int, str_numbers))

list_int = [int(i) for i in str_numbers] # efficient list comprehensive method
print(int_numbers)
print(list_int)
```

```python
words = ["apple", "banana", "cherry"]
upper_word = list(map(str.capitalize, words))
print(upper_word)
```

```python
def get_name(person):
  return f"{person["name"]} : {person["age"]}" 

people = [{
  "name": "Harshita",
  "age": 24
},
{
  "name": "Harsh",
  "age": 16
}]

print(list(map(get_name, people)))
```
