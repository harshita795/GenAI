```python
def even_or_odd(num):
  """The function finds even or odd"""
  if num % 2  == 0:
    return "The number is even"
  else:
    return "The number is odd"

print(even_or_odd(12))
```

```python
# function with multiple parameters
def add(a,b):
  """The function add two parameters"""
  return a + b

print(add(5,8))
```

```python
# Default parameters
def greet(name):
  """This functiion will greet"""
  return f"Hello {name}!"

print(greet("Harshita!"))

def greet_me(name = "Guest"):
  """This functiion will greet"""
  return f"Hello {name}!"

print(greet_me()) # <- first check here, if not then took default parameter
```

```python
# Variable length arguments
# Positional and keywords arguments

def print_numbers(*Harshita):
  """This function will print numbers"""
  for number in Harshita:
    print(number)
  
print_numbers(1,2,3,4,5)

# best practice of positonal arguments
def print_numbers(*args):
  """This function will print numbers"""
  for number in args:
    print(number)
  
print_numbers(1,2,3,4,5)
```

```python
# keywords argument, all params will be in form of key-value pairs
def print_details(**kwargs):
  """This function will print details"""
  for key,value in kwargs.items():
    print(f"{key}:{value}")

print_details(name="Harshita",age=24)

  
```

```python
# Combine
def print_details(*args,**kwargs):
  """This function will print number and details"""
  for number in args:
    print(f"Positional argument =>{number}")

  for key,value in kwargs.items():
    print(f"Keyword Argument => {key}:{value}")

print_details(1,2,3,4,5,"hello",name="Harshita",age=24)
```

```python
# Return statement
def return_statement(a,b):
  "This will let you know what retun means"
  return a * b, a, b,a ,b

print(return_statement(3,4))
```
