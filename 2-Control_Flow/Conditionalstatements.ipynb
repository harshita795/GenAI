# Conditional Statements (if, elif, else)

- Conditional Statements
- if Statement
- else Statement
- elif Statement

```python
# if , else statement
age = 24
if (age < 20):
  print("good to live")
else:
  print("good to work hard")
```

```python
#  elif statement

age = 80
if (age < 18):
  print("Very young")
elif (age > 18 and age < 25):
  print("Working")
elif (age > 25 and age < 35):
  print("Senior")
elif (age > 35 and age < 45):
  print("Parents")
elif (age > 45 and age < 55):
  print("Old")
elif (age > 55 and age < 65):
  print("Very Old")
else:
  print("Near to death")
```

```python
# Nested conditional statement

age = 26
if age >= 21 and age <= 25:
    print("Age is greater than 20")
    if age == 25:
      print("Age is 25")
  
elif age >= 11 and age <= 20:
  print("Age is between 10 to 20")
else:
  print("Age is greater than 25.")
```

```python
# examples

# Find a leap year using nested conditions

# A year is a leap year if it is divisible by 4
# However, there is an exception for century years (years ending in 00, like 1900, 2000, 2100). Century years are only leap years if they are also divisible by 400.
# 2000 is a leap year (divisible by 400).
# 1900 is NOT a leap year (not divisible by 400).

year = 2024

if year % 4 == 0:
  if year % 100 == 0:
    if year % 400 == 0:
      print(year, "The year is a leap year.")
    else:
      print(year, "is Not a leap year.")

  else:
    print(year, "is a leap year.")

else:
  print(year, "Not a leap year.")

```

```python
year = 2000
if (year % 4 == 0 and year % 100 != 0 ) or year % 400 == 0:
  print(year, "Year is leap.")
else:
  print(year, "Year is not leap.")
```

```python
# Assignment
# Calculator
num1 = float(input("Enter num1"))
num2 = float(input("Enter num2"))
operation = input("Enter operation (+, -, *, /)")
 
if operation == '+':
  result = num1 + num2
elif operation == "-":
  result = num1 - num2
elif operation == "*":
  result = num1 * num2
elif operation == "/":
  if num2 != 0:
    result = num1 / num2
  else: 
    result = "Error! Division by zero"
else:
  print("Invalid operator")

print(result)
  

```

```python
# Determine the ticket price based on age and whether the person is a student.

age = int(input("Enter the age: "))
is_student = bool(input("Enter are you a student as True or False"))

if age <= 10:
  if is_student:
    price = "$10"
  else: 
    price = "$15"
elif age > 10 and age <= 20:
  if is_student:
    price = "$22"
  else: 
    price = "$25"
elif age > 20 and age < 30:
  if is_student:
    price = "$32"
  else: 
    price = "$35"
else:
  price = "$50"

print(price)
```
