- break

```python
#  for loop

for i in range(1, 8): # last number is not included in range
  print(i)
```

```python
# third argument in range is for increment by step size
for j in range(1, 8, 2):
  print(j)
```

```python
for i in range( 10, 1, -2):
  print(i)
```

```python
# strings

str = "Harshita"
for i in str:
  print(i)
```

```python
# while loop

# while loop continues to execute as long as the condition is True.

count = 0

while count < 5:
  print(count)
  count = count + 1
```

```python
count = 2

while count % 2 == 0:
  print(count)
  count += 1
```

```python
#  loop control statements

# break
# the break statement exits the loop prematurely.

for i in range(10):
  if i == 5:
    break
  print(i)

```

```python
# continue
#  It skip the current iteration and continue with the next

for i in range(10):
  if i >= 5 and  i <= 7:
    continue
  print(i)
```

```python
# pass

# We use pass as a temporary placeholder when Python requires code to be written, but you don't want it to do anything yet.

# It is null operator, it does nothing.

for i in range(5):
  pass
  print(i)
```

```python
# Nested loops

# A loop inside a loop
for i in range(2):
  for j in "Harshi":
    print(f"i:{i} and j:{j}")
```

```python
# examples
# Calculate the sum of first N natural numbers using a while and for loop

#  while loop
sum = 0
count = 1
while count <= 9:
  sum += count
  count += 1
print(sum)

#  for loop
sum = 0
for i in range(10):
  sum += i
print(sum)
```

```python
# Examples
# Display prime numbers between 1 and 100

for num in range(1, 101):
  if num > 1:
    for i in range(2, num):
      if num % i == 0:
        break
    else:
        print(num)

    
```

```python
# fast performance,
# For example, to check if 100 is prime, you don't need to check up to 99. You only need to check up to \(\sqrt{100} = 10\). If no number up to 10 divides it, nothing will, it reduce workload.

import math

for num in range(1, 101):
  if num > 1:
    for i in range(2, int(math.sqrt(num)) + 1):
      if num % i == 0:
        break
    else:
      print(num)
```
