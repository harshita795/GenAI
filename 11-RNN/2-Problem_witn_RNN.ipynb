## Problems with RNN
- RNNs -> Sequential data -> Textual data, time series
- RNN suffers 2 major problems
  - Problem of long term dependency
  - Stagnated training


## Problem of long term dependency

- RNN forgot vary far words. This is because of Vanishing gradient problem.


### Solution
- Different activaton functions like relu, leaky relu.
- Better weight initialization.
- Skip RNNS
- LSTM

## Stagnated training
- Means RNN will not give good result, loss function will not improved, training will not able to be done.
- Reason is Explosing gradient problems.

### Solution
- Gradient Clipping.
- Controlled Learning Rate
- LSTM
