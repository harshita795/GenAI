## RNN forword propagation

-The forward propagation of a Recurrent Neural Network (RNN) computes the sequence of hidden states and outputs step-by-step through time. At each time step t, the network processes the current input \(x_{t}\) alongside the hidden state from the previous step \(h_{t-1}\) to update its memory and make a prediction.
