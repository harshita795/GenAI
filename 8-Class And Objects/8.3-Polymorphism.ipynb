- Polymorphism

Polymorphism is a core concept in Object-Oriented Programming (OOP) that allows objects of different classes to be treated as objects of a common superclass. It provides a way to perform a single action in different forms. Polymorphism is typically achieved through method overriding and interfaces

```python
# Method Overriding

# Method overriding allows a child class to provide a specific implementation of a method that is already defined in its parent class.

class Animal: # Base class
  def speak(self):
    return f"Sound of the animal"

class Dog(Animal): # Derived class
  def speak(self):
    return f"Woof!"

class Cat(Animal): # Dervied class
  def speak(self):
    return f"Meow!"

# function that demonstrate polymorphism
def animal_speak(animal):
  return animal.speak()

print(Dog().speak())
print(Cat().speak())
print(animal_speak(Dog()))
```

```python
# Polymorphism with functions and methods
# Base class
class Shape:
  def area(self):
    return f"The are of the figure."

# Derived class 1
class Rectangle(Shape):
  def __init__(self, width : float, height : float):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

# Derived class 2
class Circle(Shape):
  def __init__(self, radius : float):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius * self.radius
  
# Function that demonstrates polymorphism
def print_area(shape):
  return f"The area is {shape.area()}"

print(print_area(Circle(3)))
print(print_area(Rectangle(2, 3)))
```

```python
# Polymorphism with Abstract Base Classes

# Abstract Base Classes (ABCs) are used to define common methods for a group of related objects. They can enforce that derived classes implement particular methods, promoting consistency across different implementations.

from abc import ABC, abstractmethod

# Defien an abstract class
class Vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass

# Derived class 1
class Car(Vehicle):
  def start_engine(self):
    return f"Car engine started"

# Derived class 2
class Motorcycle(Vehicle):
  # def start_engine(self):
  #   return f"Motorcycle engine started"
  pass

# function that demostrates polymorphism
def start_vehicle(vehicle):
  return vehicle.start_engine()


# Object of car and motorcycle
print(start_vehicle(Car()))
print(start_vehicle(Motorcycle()))

# Abstract Classes (ABC) = Enforces the rules so that overriding and polymorphism are guaranteed to work safely.
```
