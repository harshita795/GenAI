- Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC, abstractmethod

# Abstract base class
class Vehicle(ABC):
  def drive(self):
    return "The vehicle is used for driving"

  @abstractmethod
  def start_engine(self):
    pass

class Car(Vehicle):
  # def start_engine(self):
    # return f"Car engine started"
  # forgot here to create start_engine method, so absrtract method will not silently crash application, but it will provide error.

 def operate_vehicle(vehicle):
  return vehicle.start_engine()

  # Abstract Methods are the parent class forcing you to write that behavior.
  # Method Overriding is you choosing to change a behavior

car = Car()
print(operate_vehicle(car))
```
