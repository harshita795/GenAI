- Inheritance In Python

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class.

```python
# Inheritance (Single Inheritance)
# Parent class
class Car:
  def __init__(self, windows : int, doors : int, engine_type : str):
    self.windows = windows
    self.doors = doors
    self.engine_type = engine_type 

  def drive(self):
    return f"The person will drive the {self.engine_type} car"

car1 = Car(4, 5, "petrol")
print(car1.drive())
```

```python
class Tesla(Car):
  def __init__(self, windows : int, doors : int, engine_type : str, is_self_driving : bool):
    super().__init__(windows, doors, engine_type)
    self.is_self_driving = is_self_driving

  def self_driving(self):
    return f"Tesla is {self.engine_type} supports self driving : {self.is_self_driving}"

tesla1 = Tesla(4, 5, "electric", True)
print(tesla1.self_driving())
print(tesla1.drive())
```

```python
# Multiple Inheritance
# When a class inherits from more than one base class
class Animal: # Base class 1
  def __init__(self, name : str):
    self.name = name

  def speak(self):
    return f"Subclass must implement this method"

class Pet: # Base class 2
  def __init__(self, owner : str):
    self.owner = owner

class Dog(Animal, Pet): # Derived class
  def __init__(self, name : str, owner : str):
    Animal.__init__(self, name)
    Pet.__init__(self, owner)

  def speak(self): # Method overriding
    return f"{self.name} says woof of the owner : {self.owner}"

dog1 = Dog("Buddy", "Harshita")
print(dog1.speak())
```
