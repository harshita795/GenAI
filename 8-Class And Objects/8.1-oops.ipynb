- Classes and Objects

Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects. 

```python
class Car:
  pass

audi = Car()
bmw = Car()

print(type(audi))
print(audi)

audi.windows = 4
print(audi.windows)

tata = Car()
tata.doors = 4
print(tata.doors)
```

```python
# Instance Variable and Methods
class Dog:
  # constructor
  def __init__(self, name = str, age = int):
    self.name = name
    self.age = age

  # objects
dog1 = Dog("tommy", 5)
print(dog1)
print(dog1.name)
print(dog1.age)
dog2 = Dog("Lucy", 7)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
# Instance method
class Dog:
  def __init__(self, name = str, age = int):
    self.name = name
    self.age = age

  def bark(self):
    print(f"{self.name} says woof")

dog1 = Dog("Tommy", 3)
dog2 = Dog("Lucy", 7)
dog1.bark()
dog2.bark()
```

```python
# Modelling a Bank Account
class Bank_Account():
  def __init__(self, owner = str, balance = 0):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    return f"{amount} is deposited. New balance is {self.balance}"
  
  def withdraw(self, amount):
    if amount > self.balance:
      return f"Insufficient find!"
    else:
      self.balance -= amount
      return f"{amount} is withdrawn. New Balance is {self.balance}"
  
  def get_balance(self):
    return self.balance

account1 = Bank_Account("Harshita", 1000)
print(account1.deposit(500))  
print(account1.withdraw(200))
print(account1.get_balance())  
```
